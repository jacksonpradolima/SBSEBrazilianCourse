## Objetivo
Demonstrar ao aluno técnicas de otimização de pesquisa para resolver problemas em teste de software.

### Objetivos Específicos
Ao final do curso, os alunos serão capazes de:
- Entender os conceitos fundamentais e técnicas de SBSE;
- Identificar diferentes tipos de técnicas de SBSE e suas aplicações;
- Criar, implementar e avaliar soluções de SBSE;
- Aplicar técnicas de SBSE para resolver problemas de engenharia de software;
- Analisar o impacto de SBSE nas práticas e processos de SBSE.
