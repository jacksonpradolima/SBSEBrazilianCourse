## Objetivo

O objetivo deste trabalho é capacitar o discente a compreender a relação entre a parte prática e teórica do tema **"Search-Based Software Engineering"** e sua relação com a indústria. Para isso, o trabalho é organizado em dois subprojetos com enfoque em:
1.	Automatização na criação de testes para sistemas baseados em Java;
2.	Priorização de casos de teste em integração contínua.

O discente irá realizar a aplicação e avaliação para ambos os subprojetos. Sendo que, o discente deverá alterar a abordagem do segundo subprojeto e avaliar o seu impacto. Ao final, o(s) artefatos(s) utilizado(s) e um relatório para cada subprojeto deverão ser entregues e apresentados.
 
## Definição

### Trabalho em Grupo 1 (TG1): Automatização na criação de testes para sistemas baseados em Java

A utilização de ferramentas e bibliotecas para a geração automática de dados de teste é uma atividade prática importante para aprender a aplicar as técnicas aprendidas de forma mais eficiente. Ferramentas e bibliotecas como [EvoSuite](https://www.evosuite.org/) podem simplificar o processo de geração de dados de teste e permitir que o discente se concentre na análise e interpretação dos resultados.

Neste subprojeto, o discente deve seguir os passos abaixo:

1. **Instalar e configurar o [EvoSuite](https://www.evosuite.org/)**: Instale a ferramenta e configure-a de acordo com as instruções fornecidas pela documentação da ferramenta.
2. **Estudar a documentação e exemplos**: Familiarize-se com a ferramenta estudando a documentação e analisando exemplos de uso. Isso ajudará o discente a entender como aplicar a ferramenta em um projeto de software.
3. **Aplicar a ferramenta em um projeto de software**: Escolha um projeto de software real aplique a ferramenta para gerar dados de teste automaticamente. Siga as diretrizes e recomendações fornecidas pela documentação para obter os melhores resultados.
4. **Analisar os resultados obtidos**: Examine os dados de teste gerados e determine se eles são eficazes na cobertura de código e detecção de defeitos. Considere ajustar os parâmetros da ferramenta ou modificar a configuração, se necessário, para obter melhores resultados.

### Trabalho em Grupo 2 (TG2): Priorização de casos de teste em integração contínua 

Comumente, empresas precisam se reinventar para estarem competitivas no mercado. Por exemplo, ChatGPT e seu impacto na comunidade. Nesse sentido, utilizando abordagens, bibliotecas e ferramentas disponíveis para melhoria de processos e produtos. O segundo subprojeto visa analisar uma abordagem do estado da arte para teste de regressão denominada [COLEMAN](https://github.com/jacksonpradolima/coleman4hcs), considerando um tópico em alta na indústria (integração contínua), e alterá-la de modo a obter uma possível melhoria nos resultados. Assim, provendo uma experiência da conexão entre indústria e academia, os quais andam lado a lado. 

Neste subprojeto, o discente deve seguir os passos abaixo:

1. **Instalar e configurar o [COLEMAN](https://github.com/jacksonpradolima/coleman4hcs)**: Instale a abordagem e configure-a de acordo com as instruções fornecidas pela documentação da abordagem.
2. **Estudar a documentação e exemplos**: Familiarize-se com a abordagem estudando a documentação, artigos relacionados, e analisando exemplos de uso. Isso ajudará o discente a entender como aplicar a abordagem em um projeto de software.
3. **Alterar a abordagem**: Analise a abordagem e proponha alterações visando a melhoria dos resultados para o problema que a abordagem visa a solucionar: priorização de casos de tese em ambiente de integração contínua. 
4. **Analisar os resultados obtidos**: Examine os resultados gerados e determine se eles são superiores aos resultados da abordagem sem alterações. Considere ajustar os parâmetros da abordagem ou modificar a configuração, se necessário, para obter melhores resultados.

# Projeto da Disciplina - Recuperação

## Objetivo
O principal objetivo deste projeto é desenvolver uma abordagem simples para aprimorar o planejamento de sprints no desenvolvimento ágil de software, utilizando técnicas de Engenharia de Software Baseada em Busca (_Search-Based Software Engineering, SBSE_). O foco será encontrar alocações de tarefas adequadas que equilibrem a carga de trabalho entre os membros da equipe e otimizem o tempo de conclusão do sprint. 

## Contexto
SBSE é uma área emergente que combina algoritmos de busca com tarefas de engenharia de software para otimizar a qualidade e os processos de desenvolvimento de software. No desenvolvimento ágil de software, o planejamento de sprints é uma atividade crucial que envolve a estimativa da complexidade das tarefas e a alocação de tarefas aos membros da equipe. Um planejamento de sprint eficaz pode levar a uma maior produtividade da equipe e melhores resultados no projeto.

## Descrição do Projeto
Este projeto consistirá nas seguintes etapas:
1.	**Revisão da Literatura**: Realizar uma revisão básica da literatura sobre técnicas de SBSE, desenvolvimento ágil de software e algoritmos de busca relevantes para o projeto. A revisão deve fornecer uma base para o projeto e ajudar a identificar algoritmos de busca adequados para a otimização do planejamento de sprints.
2.	**Formulação do Problema**: Definir o problema de otimização em termos de objetivos e variáveis de decisão. Os objetivos podem incluir minimizar o tempo de conclusão do sprint e equilibrar a carga de trabalho entre os membros da equipe. As variáveis de decisão podem ser a alocação de tarefas aos membros da equipe.
3.	**Seleção e Implementação do Algoritmo**: Selecione um algoritmo de busca apropriado (por exemplo, Algoritmo Genético, NSGA-II) que possa lidar com o problema de otimização. Implemente o algoritmo escolhido em uma linguagem de programação compatível com os requisitos do projeto.
4.	**Desenvolvimento e Avaliação de Casos de Teste**: Desenvolva casos de teste que representem diferentes cenários de planejamento de sprint com complexidades de tarefa e tamanhos de equipe variados. Aplique o algoritmo implementado nesses casos de teste e avalie seu desempenho com base nos objetivos identificados.
5.	**Documentação e Apresentação**: Documente todo o processo, incluindo revisão da literatura, formulação do problema, seleção e implementação do algoritmo e desenvolvimento e avaliação de casos de teste. Prepare um relatório final e uma apresentação gravada (um vídeo) para mostrar as descobertas e contribuições do projeto para o campo do desenvolvimento ágil de software.

## Resultados Esperados
Ao concluir este projeto, os seguintes resultados são esperados:

1.	Uma compreensão básica das técnicas de SBSE e sua aplicação na otimização do planejamento de sprints para o desenvolvimento ágil de software.
2.	Um algoritmo de busca implementado que possa aprimorar o planejamento de sprints, equilibrando a carga de trabalho entre os membros da equipe e otimizando o tempo de conclusão do sprint.
3.	Um conjunto de casos de teste que demonstre a eficácia do algoritmo implementado em diferentes cenários de planejamento de sprint.
4.	Um relatório final e apresentação que destacam as descobertas e contribuições do projeto para o campo do desenvolvimento ágil de software.
